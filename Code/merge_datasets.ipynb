# Merging Demographic Datasets with Disease & EPA Dataset

```python
#first merging demographic data 
merge_1 = pd.merge(df_Andrew, df_demo, on = ['FIPS', 'Year'], how = 'outer')
merge_1
```

```python
#merge w smoker data 

merge_2 = pd.merge(merge_1, df_smoke, on = ['FIPS', 'Year'], how = 'outer')
merge_2
```

```python
merge_final = pd.merge(merge_2, df_temp, on = ['FIPS', 'Year'], how = 'outer')
merge_final
```
